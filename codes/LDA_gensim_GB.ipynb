## LDA

```python
from gensim.corpora import Dictionary 
from gensim.models import LdaModel
import gensim
import pandas as pd
```

```python
df = pd.read_csv('../datasets/total_imputing_targets_data.csv')
df[:2]
```

```python
condition_positive =  df['sentiment'] == '긍정' # 긍정
condition_negative = df['sentiment'] == '부정'  # 부정
```

```python
positive = df[condition_positive]
negative= df[condition_negative]
```

```python
# negative = pd.read_csv('../datasets/review_negative.csv')
# positive = pd.read_csv('../datasets/review_positive.csv')
```

```python
negative['preprocess_review'] = negative['preprocess_review'].astype(str)
positive['preprocess_review'] = positive['preprocess_review'].astype(str)
```

```python
negative_review = negative['preprocess_review'].apply(lambda x: x.split()).values
positive_review = positive['preprocess_review'].apply(lambda x: x.split()).values
```

```python
dictionary = Dictionary(negative_review) #부정
dictionary
```

```python
dictionary_pos = Dictionary(positive_review) #긍정
dictionary_pos
```

### transform with dictionary

```python
dictionary.doc2bow(negative_review[1]) #부정
```

```python
dictionary_pos.doc2bow(positive_review[3]) #긍정
```

```python
corpus_list = [] #부정
for sentence in  negative_review :
    morphs_vector = dictionary.doc2bow(sentence)
    corpus_list.append(morphs_vector)
corpus_list
```

```python
corpus_list_pos = [] #긍정
for sentence in  positive_review :
    morphs_vector = dictionary_pos.doc2bow(sentence)
    corpus_list_pos.append(morphs_vector)
corpus_list_pos
```

```python
lda_model = LdaModel(corpus=corpus_list, id2word=dictionary, num_topics=3) #부정
```

```python
lda_model_pos = LdaModel(corpus=corpus_list_pos, id2word=dictionary_pos, num_topics=3) #긍정
```

### 부정의 최적의 토픽 단어수 찾기 

```python
from gensim.models.coherencemodel import CoherenceModel
```

```python
start_topic = 2
end_topic = 10
coherence_scores = []
perplexity_scores = []
for topic_number in range(start_topic, end_topic+1 ) : 
    lda_model = LdaModel(corpus=corpus_list, id2word=dictionary, num_topics=topic_number)  #fit
    coherenceModel = CoherenceModel(model=lda_model, texts=negative_review, dictionary=dictionary)
    coherence_scores.append(coherenceModel.get_coherence())  #일관성 점수
    perplexity_scores.append(lda_model.log_perplexity(corpus_list)) #혼잡도 점수
```

```python
df_scores = pd.DataFrame([coherence_scores, perplexity_scores])
df_scores.T

# 혼잡도가 적고, 일관성이 높은것을 찾으면 됨 
# 최적의 인덱스 1 실제로는 num_topics=3 즉 3개의 토픽 
## 부정의 토픽의 갯수는 3개 
```

### 긍정의 최적의 토픽 단어수 찾기

```python
start_topic = 2
end_topic = 10
coherence_scores = []
perplexity_scores = []
for topic_number in range(start_topic, end_topic+1 ) : 
    lda_model = LdaModel(corpus=corpus_list_pos, id2word=dictionary_pos, num_topics=topic_number)  #fit
    coherenceModel = CoherenceModel(model=lda_model_pos, texts=positive_review, dictionary=dictionary_pos)
    coherence_scores.append(coherenceModel.get_coherence())  #일관성 점수
    perplexity_scores.append(lda_model.log_perplexity(corpus_list_pos)) #혼잡도 점수
```

```python
df_scores = pd.DataFrame([coherence_scores, perplexity_scores])
df_scores.T

# 혼잡도가 적고, 일관성이 높은것을 찾으면 됨 
# 최적의 인덱스 0 실제로는 num_topics=2 즉 2개의 토픽 
## 긍정의 토픽의 갯수는 2개 
```
