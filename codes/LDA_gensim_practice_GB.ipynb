## LDA

```python
from gensim.corpora import Dictionary 
from gensim.models import LdaModel
import gensim
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/preprocessed_total_data.csv')
negative = pd.read_csv('../datasets/review_negative.csv')
positive = pd.read_csv('../datasets/review_positive.csv')
df[:2]
```

```python
df['preprocess_review'] = df['preprocess_review'].astype(str)
df['preprocess_review'] = df['preprocess_review'].apply(lambda x: x.split())
df['preprocess_review'] = df['preprocess_review'].apply(lambda x: ' '.join(x))
```

```python
preprocess_review = df['preprocess_review'].apply(lambda x: x.split()).values
preprocess_review
```

```python
dictionary = Dictionary(preprocess_review)
dictionary
```

#### transform with dictionary

```python
dictionary.doc2bow(preprocess_review[2])
```

```python
corpus_list = []
for sentence in  preprocess_review :
    morphs_vector = dictionary.doc2bow(sentence)
    corpus_list.append(morphs_vector)
corpus_list
```

```python
lda_model = LdaModel(corpus=corpus_list, id2word=dictionary, num_topics=3)
#벡터화된 숫자와, dictionary를 함께 넣어줘서 숫자와 실제 단어를 연결하여 알수 있게끔
```

### 최적의 토픽 단어 수

```python
## 일관성 점수
from gensim.models.coherencemodel import CoherenceModel
coherenceModel = CoherenceModel(model=lda_model, texts=preprocess_review, dictionary=dictionary)
coherenceModel.get_coherence() 

# num_topics=3에 관해서 일관성이 높으냐 적으냐에 대한 점수 
# 0.5452611131296065
```

```python
## 혼잡도 점수 
lda_model.log_perplexity(corpus_list)

-7.089082972639157
```

### 최적의 토픽 단어 수 찾기

```python
start_topic = 2
end_topic = 10
coherence_scores = []
perplexity_scores = []
for topic_number in range(start_topic, end_topic+1 ) : 
    lda_model = LdaModel(corpus=corpus_list, id2word=dictionary, num_topics=topic_number)  #fit
    coherenceModel = CoherenceModel(model=lda_model, texts=preprocess_review, dictionary=dictionary)
    coherence_scores.append(coherenceModel.get_coherence())  #일관성 점수
    perplexity_scores.append(lda_model.log_perplexity(corpus_list)) #혼잡도 점수
```

```python
df_scores = pd.DataFrame([coherence_scores, perplexity_scores])
df_scores.T

# 혼잡도가 적고, 일관성이 높은것을 찾으면 됨 
# 최적의 인덱스 0 실제로는 num_topics=2 즉 2개의 토픽 
```
