##### 불용어 처리 

```python
df_stopwords = pd.read_excel('../datasets/stopword_list_MJ.xlsx')
stopwords = df_stopwords[:5].values
stopwords
```

```python
# 토크나이징 함수 정의
def tokenizer(raw, pos=["Noun","Alpha","Verb","Adjective", "Adverb", "Foreign", "KoreanParticle"], stopword=stopwords):
    tokens = [
        word for word, tag in okt.pos(
            raw, 
            norm=True,   # normalize 그랰ㅋㅏ -> 그래ㅋㅋ
            stem=True    # stemming 바뀌나->바뀌다
            )
            if tag in pos and word not in stopword
        
    ]
    return ' '.join(tokens) 
```

```python
df_skinnylab_drop_drop['tokenized_withoutstopwords_review'] = df_skinnylab_drop_drop['review_content'].apply(tokenizer)
df_skinnylab_drop_drop
```

```python
df_skinnylab_drop_drop.to_csv('../datasets/skinnylab_reviews_tokenized_withoutstopwords.csv', index=None)
```

```python
tokenized_review_list = ' '.join(list(df_skinnylab_drop_drop['tokenized_review']))
tokenized_review_list
```

```python
tokenized_review_list_morphs = okt.morphs(tokenized_review_list) #형태소 분석
# tokenized_review_list.split()  #문자열 분할
df_tokenized_review_list_morphs = pd.DataFrame(tokenized_review_list_morphs)
df_tokenized_review_list_morphs 
```
