#### 자연어 전처리

```python
review_content_list = df_skinnylab_drop_drop['review_content'].tolist()
len(review_content_list)
review_content_list
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
morphs_list_one = okt.morphs(review_content_list[1]) #토큰화
morphs_list_one
```

```python
pos_list_one = okt.pos(review_content_list[1])
pos_list_one
```

```python
pos_list_one_noJosa = [word for word, pos in pos_list_one if pos not in ['Josa', 'Punctuation', 'Number']]
pos_list_one_noJosa
```

```python
#품사 리스트에 담기 
review_content_list_pos = list()
for review_content in review_content_list:
    content_list_pos = okt.pos(review_content)
    review_content_list_pos.append(content_list_pos)
review_content_list_pos
```

```python
#형태소 리스트에 담기
review_content_list_morphs = list()
for review_morphs in review_content_list:
    review_list_morphs = okt.morphs(review_morphs)
    review_content_list_morphs.append(review_list_morphs)
review_content_list_morphs
```

```python
#'Josa', 'Punctuation', 'Number' 제거 
review_content_list_without_stop = list()
review_content_pos_without_stop = list()

for review_content_pos in review_content_list_pos:
    for (word, pos) in review_content_pos:
        if pos not in ['Josa', 'Punctuation', 'Number']:
            review_content_pos_without_stop.append(word)
            review_content_list_without_stop.append(review_content_pos_without_stop)
            
review_content_list_without_stop
```
