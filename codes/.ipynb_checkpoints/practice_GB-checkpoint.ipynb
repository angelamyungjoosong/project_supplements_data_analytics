```python
stop_words = pd.read_csv('../datasets/koreanStopwords.txt', header=None, encoding ='utf-8')
#stop_words = stop_words[0]
stop_words= stop_words[0].values
```

```python
from konlpy.tag import Okt
okt = Okt()
from mecab import MeCab
mecab = MeCab()

# í† í¬ë‚˜ì´ì§• í•¨ìˆ˜ ì •ì˜
def tokenizer(raw, pos=["Noun","Alpha","Verb","Number","Adjective"], stopword=stop_words):
    tokens = [
        word for word, tag in okt.pos(
            raw, 
            norm=True,   # normalize ê·¸ë°ã…‹ã… -> ê·¸ë˜ã…‹ã…‹
            stem=True    # stemming ë°”ë€Œë‚˜->ë°”ë€Œë‹¤
            )
            if len(word) > 1 and tag in pos and word not in stopword

    ]
    return ' '.join(tokens)  # ê²°ê³¼ ë¦¬ìŠ¤íŠ¸ë¥¼ ë¬¸ìì—´ë¡œ ê²°í•©í•˜ì—¬ ë°˜í™˜
```

```python
## í•¨ìˆ˜ì ìš© 
df_dropna['tokenized_review'] = df_dropna['review_content'].apply(tokenizer)
df_dropna[:2]
```

```python
# í™•ì¸ìš© 
test_sentence = "ì´ ë¬¸ì¥ì€ ë¶ˆìš©ì–´ ì²˜ë¦¬ê°€ ë˜ì–´ì•¼ í•©ë‹ˆë‹¤. ê°€ë‹¤ í•˜ë‹¤ ìˆë‹¤ ë˜ë‹¤ ë¨¹ë‹¤"
tokens_test = tokenizer(test_sentence)
print(tokens_test)
```

```python
def replace_word(tokenized_review):
    replaced_review = str()
    for i in range(len(df_replace_list['before_replacement'])):
        try:
          # ì¹˜í™˜í•  ë‹¨ì–´ê°€ ìˆëŠ” ê²½ìš°ì—ë§Œ ë°ì´í„° ì¹˜í™˜ ìˆ˜í–‰
          if df_replace_list['before_replacement'][i] in tokenized_review:
            tokenized_review = tokenized_review.replace(df_replace_list['before_replacement'][i], df_replace_list['after_replacement'][i])
        except Exception as e:
            print(f"Error ë°œìƒ / ì—ëŸ¬ëª…: {e}")
    return tokenized_review
```
