### APPLY 사용해서 BMI_RESULT, BMI_TYPE, GENDER_UID 컬럼 생성


```python
def bmi_result(row):
    weight = row['WEIGHT']
    height = row['HEIGHT'] / 100
    bmi = round(weight / (height * height), 2)
    return bmi
```

```python
df_bmi['BMI_RESULT'] = df_bmi.apply(bmi_result, axis=1)
```

```python
def bmi_type(BMI_RESULT):
    if (BMI_RESULT < 18.5) :  
        x = '저체중'
    elif (18.5 <= BMI_RESULT < 23) :
        x = '정상'
    elif (23 <= BMI_RESULT < 25) :
        x = '과체중'
    else :
        x = '비만'
    return x 
```

```python
df_bmi['BMI_TYPE']= df_bmi['BMI_RESULT'].apply(bmi_type)
```

```python
# def gender(sex):
#     if sex == 1:
#         gen = '남자'
#     elif sex == 2:
#         gen = '여자'
#     return gen 
# #1:남자, 2:여자 
```

```python
# df_bmi['GENDER']= df_bmi['SEX'].apply(gender)
```

```python
def gender_uid(sex):
    if sex == 1:
        gen = 'G-01'
    elif sex == 2:
        gen = 'G-02'
    return gen 
#1:남자, 2:여자 
```

```python
df_bmi['GENDER_UID']= df_bmi['SEX'].apply(gender_uid)
```

```python
df_bmi
```

```python
df_bmi['USER_ID']= df_bmi['SEX'].apply(lambda row: str(uuid.uuid4()))
```
