### 벡터 토큰화(단어 숫자화)

```python
review_list = (df_with_target['preprocess_review']).tolist()
review_list
```

```python
from sklearn.feature_extraction.text import  TfidfVectorizer
```

```python
#교육
tfidfVectorizer = TfidfVectorizer(use_idf=True)
tfidfVectorizer.fit(review_list)
```

```python
tfidfVectorizer.idf_
```

```python
feature = tfidfVectorizer.transform([review_list[3]])
feature.toarray()
```

```python
#설명변수 
features = tfidfVectorizer.transform(review_list)
features
```

```python
sentiment_list = (df_with_target['sentiment']).tolist()
sentiment_list
```

### 목표변수 생성

```python
target = sentiment_list
```

### 머신러닝 학습

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(features.toarray(), target)
```
