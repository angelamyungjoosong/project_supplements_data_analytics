### 데이터 전처리 

```python
import pandas as pd
```

```python
df_skinnylab = pd.read_csv("../data/project_supplements_data_analysis.skinnylab_reviews.csv")
df_skinnylab 
```

```python
df_skinnylab.info()
```

```python
df_skinnylab_drop  = df_skinnylab.drop(columns =['_id'])
df_skinnylab_drop
```

#### 중복값 및 null값 제거 

```python
df_skinnylab_drop = df_skinnylab_drop.drop_duplicates()
df_skinnylab_drop
```

```python
df_skinnylab_drop.isnull().sum()
```

```python
df_skinnylab_drop_drop = df_skinnylab_drop.dropna(subset=['review_name'])
```

```python
df_skinnylab_drop_drop.isnull().sum()
```

```python
df_skinnylab_drop_drop.to_csv("../data/project_supplements_data_analysis.skinnylab_reviews_drop_duplicates.csv")
```
